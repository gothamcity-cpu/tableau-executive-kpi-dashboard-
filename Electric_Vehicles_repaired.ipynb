```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year_counts = df_clean["Model Year"].value_counts().sort_index()

plt.figure()
plt.plot(year_counts.index, year_counts.values, marker="o")
plt.title("EV Count by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Vehicle Count")
plt.grid(True)
plt.show()
```

```python
top_makes = df_clean["Make"].value_counts().head(10)

plt.figure()
plt.bar(top_makes.index, top_makes.values)
plt.title("Top 10 EV Makes")
plt.xlabel("Make")
plt.ylabel("Vehicle Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
year_counts = df_clean["Electric Range"].value_counts().sort_index()

plt.figure()
plt.plot(year_counts.index, year_counts.values, marker="o")
plt.title("Electric Range")
plt.xlabel("Electric Range")
plt.ylabel("Vehicle Count")
plt.grid(True)
plt.show()
```

```python
top_cities = df_clean["City"].value_counts().head(10)

plt.figure()
plt.bar(top_cities.index, top_cities.values)
plt.title("Top 10 Cities by EV Count")
plt.xlabel("City")
plt.ylabel("EV Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
%pip install seaborn

import seaborn as sns

range_by_make = (
    df_clean.groupby("Make")["Electric Range"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure()
sns.barplot(data=range_by_make, x="Make", y="Electric Range")
plt.title("Top 10 Makes by Avg Electric Range")
plt.xlabel("Make")
plt.ylabel("Avg Electric Range (miles)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
